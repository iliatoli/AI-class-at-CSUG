```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('Mod4CT1.jpg', 0)  # Read as grayscale

# Apply filters
filter_types = ['Mean', 'Median', 'Gaussian (sigma=1)', 'Gaussian (sigma=2)']
kernel_sizes = [3, 5, 7]
sigmas = [1, 2]

plt.figure(figsize=(12, 8))

for i, kernel_size in enumerate(kernel_sizes):
    for j, filter_type in enumerate(filter_types):
        if 'Mean' in filter_type:
            filtered = cv2.blur(image, (kernel_size, kernel_size))
        elif 'Median' in filter_type:
            filtered = cv2.medianBlur(image, kernel_size)
        else:
            sigma = sigmas[0] if 'sigma=1' in filter_type else sigmas[1]
            filtered = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)

        plt.subplot(len(kernel_sizes), len(filter_types), i * len(filter_types) + j + 1)
        plt.imshow(filtered, cmap='gray')
        plt.axis('off')
        plt.title(f'{filter_type} ({kernel_size}x{kernel_size})')

plt.tight_layout()
plt.show()
```
