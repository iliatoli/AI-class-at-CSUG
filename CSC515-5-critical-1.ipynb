```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def fingerprint_enhancement(image_path):
    # Read the fingerprint image
    img = cv2.imread(image_path, 0)  # Read as grayscale

    # Define structuring element for morphological operations
    kernel = np.ones((5, 5), np.uint8)

    # Perform morphological operations (dilation, erosion, opening, and closing)
    dilated_img = cv2.dilate(img, kernel, iterations=1)
    eroded_img = cv2.erode(img, kernel, iterations=1)
    opened_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    closed_img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

    # Display original and enhanced images
    titles = ['Original', 'Dilated', 'Eroded', 'Opened', 'Closed']
    images = [img, dilated_img, eroded_img, opened_img, closed_img]

    for i in range(5):
        plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])

    plt.show()

    # Save the enhanced images (optional)
    cv2.imwrite('dilated_fingerprint.jpg', dilated_img)
    cv2.imwrite('eroded_fingerprint.jpg', eroded_img)
    cv2.imwrite('opened_fingerprint.jpg', opened_img)
    cv2.imwrite('closed_fingerprint.jpg', closed_img)

# Path to the fingerprint image
image_path = 'fingerprint_identification.jpg'

# Call the enhancement function
fingerprint_enhancement(image_path)
```
